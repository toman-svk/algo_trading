```python
import yfinance as yf
import pandas as pd

start = "2010-07-01"
end = "2023-01-31"

df = yf.download("TSLA", start=start, end=end)

df
```

```python
import matplotlib.pyplot as plt
import datetime

plt.plot(df.index, df['Close'], label='TSLA')
plt.xlabel('Date')
plt.ylabel('Close Value')
plt.title('TSLA Stock Close Value from its IPO')
plt.legend()
plt.show()
```
